```python
import torch
from diffusers import PixArtAlphaPipeline
```

```python
pipe = PixArtAlphaPipeline.from_pretrained("PixArt-alpha/PixArt-XL-2-1024-MS", torch_dtype=torch.float16)
pipe = pipe.to('cuda')
```

```python
prompt = "A alpaca made of colorful building blocks, cyberpunk"
pipe(prompt).images[0]
```

```python
prompt = "Night photo of a sports car with the text 'SD3' on the side, the car is on a race track at high speed, a huge road sign with the text 'faster', cyberpunk"
pipe(prompt).images[0]
```
